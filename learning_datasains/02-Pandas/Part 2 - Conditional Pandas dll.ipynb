# Map, Apply dan Transform

```python
#Map adalah pemetaan dalam dataframe
#jadi disini kita akan mengubah nilai dari column shelf yang awalnya numerik menjadi karakter
#di dataframe 1 jadi kita rubah menjadi "bawah"
# 1 ---> bawah
# 2 ---> tengah
# 3 ---> atas
#dengan membuat dictionary key dan value dari kedua columns

df["shelf_name"] = df.shelf.map({1 : "Bawah", 2: "Tengah", 3: "Atas"})
df.head(20)
```

```python
#membuat sebuah fungsi dulu sebelum meng apply column fat yang dimana kita akan convert nilai
# Gram ke MG
def convert_g_to_mg(x):
    return 1000*x

#Membuat fungsi untuk mengganti huruf kapitar jadi kecil dengan fungsi lower
def convert_to_lower(x):
    return x.lower()
```

```python
#Apply = melakukan sesuatu pada seluruh columns
df['fat_mg'] = df.fat.apply(convert_g_to_mg)
df.head()
```

```python
#apply ke column mfr yang dimana huruf awalnya kapital menjadi huruf kecil
df['mfr_lower'] = df.mfr.apply(convert_to_lower)
df.head()
```

```python
#membuat fungsi yang sekali pakai dan cepat dibandingkan dengan def sebagai alternatif
df['protein_mg'] = df.protein.apply(lambda x: x*1000)
df.head()
```

```python
#transform untuk mendapatkan nilai min dan max 
#sample case rating yang di scaled
#yang dimana nilai rating 93 di ubah menjadi nilai max yaitu 1 sedangkan nilai min 0 
df['rating_scaled'] = df.rating.transform(lambda x: (x-x.min()) / (x.max()-x.min()))
df.head()
```

```python
#bersih-bersih columns yang kita buat tadi
df.drop(columns=["shelf_name","fat_mg","mfr_lower","protein_mg","rating_scaled"], inplace=True)
df.head()
```

# One-Hot encoding / Dummy variable

```python
#columns vitamin yang awalnya object dirubah menjadi numerik dengan one-hot encoding 
#untuk di manfaatkan ke machine learing nantinya
#bisa pakai map untuk label encoding
df = pd.get_dummies(df,columns=['vitamins'], prefix_sep="__")
df.head()
```
