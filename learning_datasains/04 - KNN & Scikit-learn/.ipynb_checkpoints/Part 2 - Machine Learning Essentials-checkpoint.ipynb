```python
#analisa data dulu misalkan survived dan Pclass ngaruh karena keselamatan dan tingkat yang dipilih
#sedangkan nama tidak berpengaruh jadi col nama di hapus, ticket ga ngaruh, 
#gender ngaruh karena cewe di dahulukan age juga ngaruh karena yang tua dan kecil di dahulukan
df.drop(columns=["Name","Ticket"],inplace=True)
```

# Quick Exploratory Data Analysis (Quick EDA)

# Persiapan data

```python
#dataset di pecah menjadi 2 bagian yaitu input = feature || target = label
#feature X 
#label y -> survived
#Pisahin data 
#X semua colum kecuali colom target yaitu survived
```

```python
X = df.drop(columns="Survived")
y = df.Survived

#jangan lupa column categorikal dirubah karena mesin hanya bisa membaca berupa angka dengan 
#encoding data menggunakan get dummies
X = pd.get_dummies(X, columns=["Pclass","Sex","Embarked"])
```

# Intro Scikit-learn 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#(n_neighbors=1) = jumlah tetangga
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)
```

# Concept 1 : Dataset Spliting

```python
illustration.train_test_split
```

```python
#data kita classification
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns="Survived")
y = df.Survived

X = pd.get_dummies(X, columns=["Pclass","Sex","Embarked"])
```

```python
X.shape
```

```python
# stratify=y, = classification selang seling berdasarkan label survived atau non
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train), knn.score(X_test, y_test)
```

```python
#dataset spliting belum cukup karena sering over fit
```

```python
demo.knn_scaling()
```

```python
illustration.knn_scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train.head()
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_scaled, y_train)
```

```python
knn.score(X_train_scaled, y_train), knn.score(X_test_scaled, y_test)
#contoh improvement karena data ada yang ga balance sedangkan knn butuh distance
```
