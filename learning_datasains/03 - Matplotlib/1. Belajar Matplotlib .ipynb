```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#linspace membuat 10 element dibagi sama rata
#sqrt akar dari X
X = np.linspace(0,10,1000)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisialisasi Figure dan Axis

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes() #ax == kordinat
ax.plot(X,y1)
ax.plot(X,y2)
```

### shortcut

```python
plt.figure(figsize=(6,6))
plt.plot(X,y1,'r-.', linewidth=0.5, label='sin')
plt.plot(X,y2,'bo', markersize=3, label='cos')
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal') #agar skalnya sama atau balance
plt.title("sin dan cos",fontsize=16)
plt.xlabel("X", fontsize=16)
plt.ylabel("Y",fontsize=16)
plt.legend(loc='lower center', ncol=2, shadow=True)
```

```python
df = pd.read_csv("circle_data.csv", delimiter=";")
df.head()
```

```python
plt.scatter(x,y)
#harus samaain skalanya agar sama dengan equal
plt.axis('equal')
```

# Text dan annotation

```python
numbers = np.random.randn(100000)
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers, bins=100, density=1);
plt.ylim(0,1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0", ha="center")
#anonotation
plt.annotate("mean=0", xy=(0,0.4),xytext=(1, 0.5), ha="center", arrowprops={'arrowstyle':'-|>'})
```

# 3D plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x,y)
#z Parabola
Z = np.sqrt(X**2 + Y**2 + 10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, antialiased=False, cmap="winter")
```
