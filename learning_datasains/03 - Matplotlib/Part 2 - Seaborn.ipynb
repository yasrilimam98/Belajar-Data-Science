```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("linear.csv")
df.head()
```

```python
#plot dulu data 1 diambil dari column tipe dan data 2 diambil tipe juga
df1 = df[df.tipe == "data1"]
df2 = df[df.tipe == "data2"]
plt.scatter(df1.X, df1.y);
plt.scatter(df2.X, df2.y);
```

# Seaborn

```python
#Lebih mudah pakai seaborn karena hanya masukin dataframenya dan untuk mengubah warnanya tinggal
#menggunakan fungsi hue berdasarkan kolomnya contohnya pada colom tipe ada 2 jenis data 1 dan 2
#dan sudah di sediakan legendnya dan enak pakai untuk analisis data pakai seaborn
plt.figure(figsize=(6,6))
sns.scatterplot("X","y",data=df, hue="tipe");
```

```python
#untuk melakukan regresinya pada plot dengan lmplot
plt.figure(figsize=(10,10))
sns.lmplot("X","y", data=df, hue="tipe");
```

```python
#histogram itu hanya untuk melihat satu baris data
plt.figure(figsize=(6,6))
sns.distplot(df1.y, bins=20);
```

# Seaborn dan Pandas DataFrame
### Statistical Summary

```python
df = pd.read_csv("cereal.csv", index_col="name")
df.head()
```

# Missing value plot

```python
#dinegasikan di (~df agar plot lebih keliatan terang)
plt.figure(figsize=(6,6))
sns.heatmap(~df.isna(), cmap='Blues', cbar=False)
```

### Correlatian heatmap

```python
#corelasi penuh warna merah dan putih tidak
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, vmax=1, annot=True, square=True, fmt=".1f")
```

### Pairplot

```python
sns.pairplot(df)
```

```python
#seleksi pada colom tertentu 
sns.pairplot(df, hue="shelf", vars=["calories","fat","rating"], kind='reg')
```

```python
#kalau mau mengjoin 2 kolum
sns.jointplot("potass","fiber", data=df, marginal_kws={"bins": 25}, kind='reg')
```

# Data categorical

### Persebaran

```python
#kategorikal plot
sns.catplot("mfr","rating", data=df)
```

```python
sns.boxplot("mfr","rating", data=df)
```

```python
sns.violinplot("mfr","rating", data=df)
```

### Count

```python
sns.catplot("mfr", data=df, kind='count')
```

```python
sns.countplot("mfr", data=df, hue="shelf")
```
