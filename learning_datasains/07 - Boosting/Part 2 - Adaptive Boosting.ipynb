### Max Depth

```python
demo.max_depth()
```

Jumlah pohon (iterasi) bisa improve model tapi tidak menyebabkan overfit. Justru ketika kita memakai depth yang lebih dalam yang dapat menyebabkan overfit karena dia dengan mudah fit dengan residualnya.

### Learning Rate

```python
demo.learning_rate()
```

untuk learning rate tidak disarankan terlalu melebihi 1 karena artinya lebih agresif dalam fitting si residual. Bisa kita lihat juga akibatnya iterasi selanjutnya harus melawannya dengan weight yang agresif juga. Walau demikian, kalau terlalu kecil learning rate nya, maka kita tidak fit si residual sama sekali

### Kombinasi max_depth dan learning_rate

```python
demo.max_depth_and_lr()
```

overfit dari `max_depth` dapat sedikit ditanggulangi menggunakan `learning_rate` yang kecil, tetapi tetap kelihatan seperti overfit, jadi hati-hati dalam memilih `max_depth`.

Kebetulan karena problem kali ini sederhana, sebenarnya decision tree dengan `max_depth=3` sudah merupakan strong learner
